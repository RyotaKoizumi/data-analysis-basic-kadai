```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs  = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axs[0,0].plot(y0)

x1 = np.linspace(0,10, 100)
y1 = np.sin(x1)
axs[0,1].plot(x1,y1)

x2 = np.random.normal(10, 2, 2000)
axs[0,2].hist(x2, bins=100)

np.random.seed(5)
x3 = np.random.normal(5, 2, 100)
y3 = np.random.normal(10, 4, len(x3))
axs[1,0].scatter(x3, y3)

plt.show()

```
